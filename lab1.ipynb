```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

##### 读取相关数据，将数据集随机分为80%训练集，20%测试集

```python
rating = pd.read_csv("/Users/JiaoFusen/Desktop/ml-latest-small/ratings.csv")
movie = pd.read_csv("/Users/JiaoFusen/Desktop/ml-latest-small/movies.csv")
ratings = pd.merge(rating, movie, on="movieId")
ratings_train, ratings_test = train_test_split(ratings, train_size=0.8)
ratings_train
```
