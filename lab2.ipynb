```python
import time
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
rating = pd.read_csv("/Users/JiaoFusen/Desktop/ml-latest-small/ratings.csv")
rating
```

```python
movie_rating = rating.loc[:, ['movieId', 'rating']]
movie_rating
```

```python
average_rating = movie_rating.groupby('movieId').mean()
average_rating
```

```python
user_rating = rating[rating.userId == 288].sort_index(by='timestamp')
user_movie = user_rating.movieId.tolist() #用户看过的电影集合
user_rating = user_rating.loc[:, ['movieId', 'rating', 'timestamp']]
user_rating
# user_rating.set_index('movieId')
# user_movie = user_rating.loc[:, ['movieId', 'rating']]
# user_movie.set_index('movieId')
```

```python
# movie_dict = average_rating.to_dict("index")
# movie_dict
user_aver_rating = average_rating.loc[user_movie]
user_aver_rating
```

```python
user_rating = user_rating.merge(user_aver_rating, right_index=True, how="outer", left_on='movieId')
```

```python
user_rating1 = user_rating.loc[:, ['rating_x', 'rating_y']]
# user_rating1.reindex(np.arange(len(user_rating1)))
index = np.arange(len(user_rating1))
user_rating1.set_index(index)
```

```python
plt.figure(figsize=(25, 25))
plt.plot(np.arange(len(user_rating1)), user_rating1.rating_x, lw=3)
plt.plot(np.arange(len(user_rating1)), user_rating1.rating_y, lw=3)
plt.xticks(np.arange(0, 80, 5))
```
