## 整合平均数和itemrank值

```python
import numpy as np
import pandas as pd
from ItemRank import ItemRank
from mean import Mean
import matplotlib.pyplot as plt
np.set_printoptions(threshold=np.inf)
plt.style.use('ggplot')
%matplotlib inline
```

### 从文件中读取数据

```python
pd_data = pd.read_csv("D:/ml-latest-small/ratings.csv")
np_data = pd_data.values
```

### 计算itemrank值

```python
item_rank = ItemRank(np_data)
```

### 生成图模型

```python
item_rank.generate_graph()
```

### 计算相关系数

```python
item_rank.generate_coef_from_graph()
```

### 生成评分向量

```python
result_all = {}
user_names = item_rank.user_names
for u in user_names:
    d = item_rank.generate_d(user_name=u)
    IR = d
    covered = False
    counter = 0
    while not covered:
        counter += 1
        old_IR = IR
        IR = item_rank.item_rank(0.85, IR, d)
        covered = (old_IR - IR < 0.0001).all()
    mean = Mean(pd_data)
    average = mean.calc_mean(u)
    movie_by_time = average.index.tolist()
    IR_by_time = []
    for m in movie_by_time:
        IR_by_time.append(IR[item_rank.movie_names.index(m)])
    IR_by_time = {'IR': IR_by_time}
    IR_by_time = pd.DataFrame(IR_by_time).set_index(pd.Series(movie_by_time))
    result = average.merge(IR_by_time, left_index=True, right_index=True, how="outer")
    result_all[u] = result
    print(result)
```

```python
import pickle
f = open('IR.pkl', 'wb')
pickle.dump(result_all, f)
f.close()
```
